This code is used to simulate an hotel where the floor is changed based on a dice roll. If the dice rolls an 1 or a 2, the elevator goes down one floor. If the result is somewhere between 3 and 5, the elevator goes up one floor, and if you roll a 6, you have to roll the dice again, and you go up the number rolled on the second roll.

```python
from random import randint

def roll_dice():
	roll = randint(1,6)
	return roll

epoch = 100
number_of_rounds = 500
rounds = []
current_roll = 0
current_floor = 50
back_to_zero = randint(1,1000)
is_zero = 0
second_roll = 0

for i in range(1, number_of_rounds):
	current_floor = 50
	for j in range(1, epoch):
		is_zero = randint(1,1000)
		if is_zero == back_to_zero:
			current_floor = 0
			rounds.append(current_floor)
		else:
			current_roll = roll_dice()
			if current_roll == 1 or current_roll == 2:
				current_floor -= 1
				rounds.append(current_floor)
			elif current_roll >= 3 and current_roll <=5:
				current_floor += 1
				rounds.append(current_floor)
			elif current_roll == 6:
				second_roll = roll_dice()
				current_floor += second_roll
				rounds.append(current_floor)

print (rounds)
```
